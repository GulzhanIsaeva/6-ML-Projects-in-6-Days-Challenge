```python
# Pie Chart
dataset2 = dataset[['housing', 'is_referred', 'app_downloaded', 'web_user', 
                    'app_web_user', 'ios_user', 'android_user', 'registered_phones',
                    'payment_type', 'waiting_4_loan', 'cancelled_loan', 'received_loan',
                    'rejected_loan', 'zodiac_sign', 'left_for_two_month_plus', 'left_for_one_month']]

fig = plt.figure(figsize=(45, 35))
plt.suptitle('Pie Chart Distributions', fontsize=40)
for i in range(1, dataset2.shape[1] + 1):
    plt.subplot(6, 3, i)
    f = plt.gca()
    f.set_title(dataset2.columns.values[i - 1])
    values = dataset2.iloc[:, i - 1].value_counts(normalize = True).values
    index = dataset2.iloc[:, i - 1].value_counts(normalize = True).index
    
    
    
    plt.pie(values, labels = index, autopct='%1.1f%%')
    plt.axis('equal')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
dataset = dataset[pd.notnull(dataset['age'])]
```

```python
dataset2 = dataset.drop(columns = ['credit_score', 'rewards_earned'])
```

```python
# Histograms
dataset2 = dataset.drop(columns = ['user', 'churn'])

fig = plt.figure(figsize=(18, 15))
plt.suptitle('Histograms of Numerical Columns', fontsize=30)
for i in range(1, dataset2.shape[1] + 1):
    plt.subplot(6, 5, i)
    f = plt.gca()
    f.set_title(dataset2.columns.values[i - 1])

    vals = np.size(dataset2.iloc[:, i - 1].unique())
    if vals >= 100:
        vals = 100
    
    plt.hist(dataset2.iloc[:, i - 1], bins=vals, color='#3F5D7D')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```
